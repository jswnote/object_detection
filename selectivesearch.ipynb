```python
import cv2
import numpy as np
```

```python
image = cv2.imread('/Users/jsw/Desktop/boksilee2.jpg', cv2.IMREAD_COLOR)
```

```python
if image is None:
    raise Exception('영상 파일 읽기 오류')
```

```python
import selectivesearch
```

```python
_, regions = selectivesearch.selective_search(image, scale = 100, min_size = 2000)
```

```python
type(regions)
```

```python
len(regions)
```

```python
regions
#rect : (x,y,w,h), label : 비슷한 종류들 끼리 합침.
```

```python
show_rect = [i['rect'] for i in regions]
```

```python
show_rect
```

```python
b, g, r = (255, 0, 0), (0, 255, 0), (0, 0, 255)
```

```python
for i in show_rect:
    cv2.rectangle(image, (i[0], i[1], i[2], i[3]), b, thickness = 2)
```

```python
cv2.imshow('image', image)
cv2.waitKey(0)
```
