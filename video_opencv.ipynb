```python
import cv2
import numpy as np
```

```python
video_input_path = '/Users/jsw/Desktop/VOCdevkit/video.mp4'
video_output_path = '/Users/jsw/Desktop/VOCdevkit/video_out.mp4'
```

```python
cap = cv2.VideoCapture(video_input_path)
codec = cv2.VideoWriter_fourcc('m','p','4','v')
```

```python
vid_size = (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
vid_fps = cap.get(cv2.CAP_PROP_FPS)
```

```python
vid_writer = cv2.VideoWriter(video_output_path, codec, vid_fps, vid_size)
```

```python
frame_cnt = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
```

```python
print('총 프레임 개수 : ', frame_cnt, 'FPS : ', round(vid_fps), 'Frame 크기 : ', vid_size)
```

```python
import time
```

```python
b, g, r = (255, 0, 0), (0, 255, 0), (0, 0, 255)
```

```python
start = time.time()
```

```python
index = 0
```

```python
while True:
    hasFrame, imgFrame = cap.read()
    if not hasFrame:
        print('더이상의 프레임은 없습니다')
        break
    
    index += 1
    print('frame', index, '처리완료')
    
    cv2.rectangle(imgFrame, (600,200,200,200), g, thickness = 2)
    caption = 'frame:{}'.format(index)
    cv2.putText(imgFrame, caption, (300, 95), cv2.FONT_HERSHEY_SIMPLEX, 0.7, r, 1)
    
    vid_writer.write(imgFrame)
    
print('write 완료시간', round(time.time()-start, 4))
vid_writer.release()
cap.release()
```
