```python
#처음부터 돌려야 이게 업데이트가 되네
```

```python
def IOU(cand_box, gt_box):
    x1 = np.maximum(cand_box[0], gt_box[0])
    y1 = np.maximum(cand_box[1], gt_box[1])
    x2 = np.minimum(cand_box[2], gt_box[2])
    y2 = np.minimum(cand_box[3], gt_box[3])
    
    box = np.maximum(x2-x1, 0) * np.maximum(y2-y1, 0) #겹친 사각형
    cand_box_area = (cand_box[2]-cand_box[0]) * (cand_box[3]-cand_box[1])
    gt_box_area = (gt_box[2]-gt_box[0]) * (gt_box[3]-gt_box[1])
    
    union = cand_box_area + gt_box_area - box
    
    iou = box / union
    
    return iou
```

```python
gt_box = [150, 300, 790, 900]
```

```python
image = cv2.imread('/Users/jsw/Desktop/boksilee2.jpg', cv2.IMREAD_COLOR)
image_rgb_cand = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
_, regions = selectivesearch.selective_search(image_rgb_cand, scale=100, min_size=2000)

print(type(regions), len(regions))
```

```python
rects = [cand['rect'] for cand in regions if cand['size'] > 5000]
```

```python
len(rects)
```

```python
for index, cand_box in enumerate(rects):
    cand_box = list(cand_box)
    cand_box[2] += cand_box[0]
    cand_box[3] += cand_box[1]
    
    iou = IOU(cand_box, gt_box)
    print('index:', index, "iou:", iou)
```

```python
image = cv2.imread('/Users/jsw/Desktop/boksilee2.jpg', cv2.IMREAD_COLOR)
image_rgb_last = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print('img shape:', image.shape)
```

```python
green_rgb = (125, 255, 51)
image_rgb = cv2.rectangle(image_rgb_last, (gt_box[0], gt_box[1]), (gt_box[2], gt_box[3]), color=red, thickness=2)

for index, cand_box in enumerate(rects):
    
    cand_box = list(cand_box)
    cand_box[2] += cand_box[0]
    cand_box[3] += cand_box[1]
    
    iou = IOU(cand_box, gt_box)
    
    if iou > 0.5:
        print('index:', index, "iou:", iou, 'rectangle:',(cand_box[0], cand_box[1], cand_box[2], cand_box[3]) )
        cv2.rectangle(image_rgb_last, (cand_box[0], cand_box[1]), (cand_box[2], cand_box[3]), color=green_rgb, thickness=1)
        text = "{}: {:.2f}".format(index, iou)
        cv2.putText(image_rgb_last, text, (cand_box[0]+ 100, cand_box[1]+10), cv2.FONT_HERSHEY_SIMPLEX, 0.4, color=green_rgb, thickness=1)
    
plt.figure(figsize=(12, 12))
plt.imshow(image_rgb_last)
plt.show()
```
